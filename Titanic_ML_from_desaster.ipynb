Seleo de Modelo

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(['Survived'], axis='columns')
y = df['Survived']
```

```python
x_treino,x_teste,y_treino,y_teste = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(len(x_treino),len(y_treino),len(x_teste),len(y_teste))
```

```python
from sklearn.linear_model import LogisticRegression
mrl = LogisticRegression()
```

```python
mrl.fit(x_treino, y_treino)
```

```python
prl = mrl.predict(x_teste)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_teste, prl))
```

```python
from sklearn import tree
mad=tree.DecisionTreeClassifier()
mad.fit(x_treino,y_treino)
```

```python
pad = mad.predict(x_teste)
print(accuracy_score(y_teste, pad))
```
